```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
```

```python
train_data = pd.read_csv('train_data.csv', header=None, index_col=0)
train_data
```

```python
test_data = pd.read_csv('test_data.csv', header=None, index_col=0)
test_data
```

```python
y = pd.read_csv('train/targets.tsv', header=None, index_col=0, sep='\t').loc[train_data.index,:]
y
```

### Обучение модели

```python
X_train, X_val, y_train, y_val = train_test_split(train_data, y, stratify=y)
```

```python
params = {'loss_function':'Logloss',
          'eval_metric':'Accuracy',
          'verbose': 200,
          'random_seed': 0,
          'iterations': 10000
         }
model = CatBoostClassifier(**params)
model.fit(X_train, y_train,
          eval_set=(X_val, y_val),
          use_best_model=True,
          plot=True
         )
```

```python
test_data['pred'] = model.predict(test_data)
test_data['pred']
```

```python
test_data['pred'].to_csv('test/targets.tsv', sep='\t', header=False)
```

### Top-25 признаков

```python
A = model.get_feature_importance()
B = A.argsort()[::-1]
feature_importances = pd.DataFrame({'feature_code': train_data.columns[B]}) 
stats = ['mean', 'var', 'median', 'max']
feature_importances['feature'] = feature_importances['feature_code'].apply(lambda x: (x - 2)%64 
                                                                           if x >= 1 else -1)
feature_importances['stat'] = feature_importances['feature_code'].apply(lambda x: stats[(x - 2)//64] 
                                                                        if x >= 1 else 'len')
feature_importances['importance'] = A[B]
feature_importances = feature_importances.drop(columns=['feature_code'])
feature_importances.head(25)
```
